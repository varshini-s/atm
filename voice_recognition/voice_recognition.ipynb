```python
import speech_recognition as sr 
import pyttsx3  
from flashtext import KeywordProcessor
# Record Audio
r = sr.Recognizer()
with sr.Microphone() as source:
    print("Say something!")
    audio = r.listen(source)
    

flag=0
 
# Speech recognition using Google Speech Recognition
try:
    # for testing purposes, we're just using the default API key
    # to use another API key, use r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")
    # instead of r.recognize_google(audio)
    print("You said: " + r.recognize_google(audio))
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
except sr.RequestError as e:
    print("Could not request results from Google Speech Recognition service; {0}".format(e))
words = {
    'help': 0,
    'emergency': 0,  
    'police': 0,
    'rescue': 0,
    'thief': 0,
    'stop': 0,
    'please': 0,
    'burglar': 0,
    'robber': 0
    }
document=r.recognize_google(audio)
processor = KeywordProcessor()
processor.add_keywords_from_list(list(words.keys()))
found = processor.extract_keywords(document)
print(found)

word = list(words.keys())
freq = []
for k in word:
    words[k] = document.count(k)
    freq.append(words[k])

  
def fun(arr):
    l = 0
    t=max(arr)
        
        
    if (t == 0):
        l = 1
    elif t >= 3:
        l = 2
         
    else:
        l =3
            
    return l

if(fun(freq)==1):
    print("no danger word detected")
elif(fun(freq)==2):
    print('keyword found')
    for k in word:
        print('frequency of word:{} is'.format(k),words[k])
    print("emergency situation")


elif(fun(freq)==3 ):
    print('keyword found')
    for k in word:
        print('frequency of word:{} is'.format(k),words[k])
   
    print("alert alert")
```
